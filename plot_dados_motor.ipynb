```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dados = pd.read_csv('dados_teste_carga.csv')
# dados = pd.read_csv('dados_1000RPM.csv')


tempo = dados['Tempo (s)']
rpm = dados['RPM']

plt.figure(figsize=(10, 6))
plt.plot(tempo, rpm, linestyle='-', color='b', label='RPM')
plt.title('Variao de RPM ao longo do Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('RPM')
plt.grid(True)
plt.legend()
plt.tight_layout()




plt.show()
```

```python
fig = px.line(dados, x='Tempo (s)', y='RPM', title='Variao de RPM ao longo do Tempo')
fig.update_xaxes(title='Tempo (s)')
fig.update_yaxes(title='RPM')
fig.show()
```

```python



window_size = 5
dados['dados_filtrados'] = dados['RPM'].rolling(window=window_size, min_periods=1).mean()


plt.figure(figsize=(10, 6))
plt.plot(dados['Tempo (s)'], dados['RPM'], label='Dados Originais')
plt.plot(dados['Tempo (s)'], dados['dados_filtrados'], label='Dados Filtrados (Mdia Mvel)')
plt.xlabel('Tempo (s)')
plt.ylabel('RPM')
plt.title('Dados Originais vs Dados Filtrados')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


fig = px.line(dados, x='Tempo (s)', y='RPM', title='Variao de RPM ao longo do Tempo')
fig.add_scatter(x=dados['Tempo (s)'], y=dados['dados_filtrados'], mode='lines', name='RPM Filtrado')
fig.update_xaxes(title='Tempo (s)')
fig.update_yaxes(title='RPM')
fig.show()
```

```python
dados_simulink = pd.read_csv("dados_simulink.csv")
print(dados_simulink.columns)
plt.figure(figsize=(10, 6))
plt.plot(dados_simulink['tempo'], dados_simulink['degrau'], label='Degrau Unitrio')
plt.plot(dados_simulink['tempo'], dados_simulink['ft_discreta_senoide'], label='Resposta com Controle Discreto com Senoide')
plt.plot(dados_simulink['tempo'], dados_simulink['ft_discreta'], label='Resposta com Controle Discreto')
plt.xlabel('Tempo (s)')
plt.ylabel('RPM')
plt.title('Simulao do Simulink para o Controlador Discretizado')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


fig = px.line(dados_simulink, x='tempo', y='degrau', title='Simulao do Simulink para o Controlador Discretizado')


fig.add_scatter(x=dados_simulink['tempo'], y=dados_simulink['ft_discreta_senoide'], mode='lines', name='ft_discreta_senoide')
fig.add_scatter(x=dados_simulink['tempo'], y=dados_simulink['ft_discreta'], mode='lines', name='ft_discreta')


fig.update_xaxes(title='Tempo (s)')
fig.update_yaxes(title='Amplitude')

fig.show()
```
