```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dados = pd.read_csv('dados.csv')
# dados = pd.read_csv('dados_1000RPM.csv')


tempo = dados['Tempo (s)']
rpm = dados['RPM']

plt.figure(figsize=(10, 6))
plt.plot(tempo, rpm, linestyle='-', color='b', label='RPM')
plt.title('Variação de RPM ao longo do Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('RPM')
plt.grid(True)
plt.legend()
plt.tight_layout()




plt.show()
```

```python
fig = px.line(dados, x='Tempo (s)', y='RPM', title='Variação de RPM ao longo do Tempo')
fig.update_xaxes(title='Tempo (s)')
fig.update_yaxes(title='RPM')
fig.show()
```

```python



window_size = 5
dados['dados_filtrados'] = dados['RPM'].rolling(window=window_size, min_periods=1).mean()


plt.figure(figsize=(10, 6))
plt.plot(dados['Tempo (s)'], dados['RPM'], label='Dados Originais')
plt.plot(dados['Tempo (s)'], dados['dados_filtrados'], label='Dados Filtrados (Média Móvel)')
plt.xlabel('Tempo (s)')
plt.ylabel('RPM')
plt.title('Dados Originais vs Dados Filtrados')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


fig = px.line(dados, x='Tempo (s)', y='RPM', title='Variação de RPM ao longo do Tempo')
fig.add_scatter(x=dados['Tempo (s)'], y=dados['dados_filtrados'], mode='lines', name='RPM Filtrado')
fig.update_xaxes(title='Tempo (s)')
fig.update_yaxes(title='RPM')
fig.show()
```
